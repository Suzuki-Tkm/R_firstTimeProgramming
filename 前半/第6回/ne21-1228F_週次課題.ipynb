###「統計テスト1」の得点が平均=20-(学生証番号末尾1桁)の正規分布であるという帰無仮説を検定せよ。

帰無仮説H:μ=12(統計テスト1の母平均は12である)

```python
data <- read.csv("/content/指導法データ.csv")
```

```python
統計テスト <- data$統計テスト１
```

```python
t分子 <- mean(統計テスト)-12
```

```python
t分母 <-sqrt(var(統計テスト)/length(統計テスト))
```

```python
t統計量 <- t分子/t分母
```

```python
t統計量
```

```python
qt(0.025,19)#自由度19のt分布で下側確率0.025となるtの値
```

```python
qt(0.975,19)#自由度19のt分布で下側確率0.975となるtの値
```

```python
qt(0.025,19,lower.tail=FALSE)#自由度19のt分布で上側確率0.025となるtの値
```

```python
pt(t統計量,19)#自由度19のt分布で下側確率probを計算
```

```python
pt(-t統計量,19,lower.tail=FALSE)#自由度19のt分布で上側確率probを計算
```

```python
2*pt(-t統計量,19,lower.tail=FALSE)#p値(自由度19)
```

p値が0.891107058898245と求められ、有意水準0.05よりも大きいので帰無仮説は棄却されない。よって統計学テストの得点が平均=20-(学生証番号末尾1桁)の正規分布である。
