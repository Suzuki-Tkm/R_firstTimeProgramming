###「心理学テスト」の得点が平均=20-(学生証番号末尾1桁)の正規分布であるという帰無仮説を検定せよ。

帰無仮説H:μ=12(心理学テストの母平均は12である)

```python
data <- read.csv("/content/指導法データ.csv")
```

```python
心理学 <- data $ 心理学テスト
```

```python
t分子 <- mean(心理学)-12
```

```python
t分母 <- sqrt(var(心理学)/length(心理学))
```

```python
t統計量 <- t分子/t分母
```

```python
t統計量
```

###棄却域を求める

```python
qt(0.025,19)#自由度19のt分布で下側確率0.025となるtの値
```

```python
qt(0.975,19)#自由度19のt分布で下側確率0.975となるtの値
```

```python
qt(0.025,19,lower.tail=FALSE)#自由度19のt分布で上側確率0.025となるtの値
```

###以上より棄却域はt<-2.09302405440831,t>2.09302405440831となる。検定統計量の実現値はt=-2.67634391942846であり、棄却域に入るので棄却される。

###p値を直接求める。

```python
pt(t統計量,19)#自由度19のt分布で下側確率probを計算
```

```python
pt(-t統計量,19,lower.tail=FALSE)#自由度19のt分布で上側確率probを計算
```

###p値を求める。

```python
2*pt(-t統計量,19,lower.tail=FALSE)#p値を計算(自由度=19)
```

###p値が0.0149312434899411と求められ、有意水準0.05よりも小さいので帰無仮説は棄却される。よって心理学テスト得点について母集団の平均が12であるとは言えない。
