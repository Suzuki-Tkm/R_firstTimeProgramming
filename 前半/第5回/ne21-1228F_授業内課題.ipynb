###2. 週次課題5.3 2.で考えたゆがんだサイコロの目の平均を求めた結果の期待値と分散を求めよ。

```python
平均 <- numeric(length=10000)
```

```python
for (i in 1:10000){
  サイコロ <- sample(1:6,10,replace=TRUE, prob=c(1/6,1/3,1/12,1/12,1/12,1/4))
  平均[i] <- mean(サイコロ)
}
```

サイコロ平均の期待値

```python
mean(平均)
```

サイコロ平均の分散

```python
var(平均)
```
