###1. 「サイコロを10回振って、その目の平均値を調べる」を10000回繰り返し、そのヒストグラムを描いてみましょう。

```python
サイコロ平均 <- numeric(length = 10000)
```

```python
for(i in 1: 10000){
  サイコロ <- ceiling(runif(n=10,min=0,max=6))
  サイコロ平均[i] <- mean(サイコロ)
}
```

```python
hist(サイコロ平均)
```

###2. 「ゆがんだサイコロを10回振って、その目の平均値を調べる」を10000回繰り返し、そのヒストグラムを描いてみましょう。

```python
平均 <- numeric(length=10000)　#10000個のサイコロの平均値を入れる配列。
```

```python
for (i in 1:10000){
  サイコロ <- sample(1:6,10,replace=TRUE, prob=c(1/6,1/3,1/12,1/12,1/12,1/4)) #1が1/6,2が1/3,3が1/12,4が1/12,5が1/12,6が1/4の歪んだサイコロ
  平均[i] <- mean(サイコロ)
}
```

```python
hist(平均)
```
