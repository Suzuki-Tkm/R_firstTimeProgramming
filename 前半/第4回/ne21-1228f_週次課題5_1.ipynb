###1. 「平均50、標準偏差10の正規分布からn=10つまり10個の数を無作為抽出して標本分散を求める」を10000回繰り返してヒストグラムを描きましょう。



```python
標本分散 <- numeric(length = 10000)
```

```python
for(i in 1:10000){
  標本 <- rnorm(n=10,mean=50,sd=10)
  標本分散[i] <- mean((標本-mean(標本))^2) 
}
```

```python
hist(標本分散)
```

###2. 「平均50、標準偏差10の正規分布からn=10つまり10個の数を無作為抽出して不偏分散を求める」を10000回繰り返してヒストグラムを描きましょう。

```python
不偏分散 <- numeric(length = 10000)
```

```python
for(i in 1:10000){
  標本 <- rnorm(n=10,mean=50,sd=10)
  不偏分散[i] <- var(標本) 
}
```

```python
hist(不偏分散)
```
