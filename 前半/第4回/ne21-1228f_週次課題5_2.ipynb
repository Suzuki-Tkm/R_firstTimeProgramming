###1. 実際にサイコロを2回だけ振って、その目の平均が3.5になる確率はどのくらいでしょうか。

```python
試行回数1 <- 50000
```

```python
平均1 <- 0
for(i in 1: 試行回数1){
  サイコロ <- ceiling(runif(n=2,min=0,max=6))
  サイコロ平均 <- mean(サイコロ)
  if(サイコロ平均 == 3.5) 平均1 <- 平均1 + 1
}
```

```python
平均1/試行回数1
```
