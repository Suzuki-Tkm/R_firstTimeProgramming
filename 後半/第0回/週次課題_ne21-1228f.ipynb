###週次課題8.1

```python
変数IDs <- score$このExcelシートは乱数を使って0から99までの整数を発生させたもので実際の成績とは一切関係ありません.
```

####grep関数を利用して次の条件に当てはまる学籍番号を取り出して下さい。

「3」を含む番号

```python
grep("\\d",変数IDs,perl=TRUE,value=TRUE)
```

```python
grep("3",変数IDs,perl=TRUE,value=TRUE)
```

チェックデジットが「E」の番号

```python
grep("E$",変数IDs,perl=TRUE,value=TRUE)
```

下4桁の番号が100番代の番号(例: NE10-1111A、NE20-0123B)

```python
grep("NE\\d{2}-\\d1\\d{2}",変数IDs,perl=TRUE,value=TRUE)
```

番号内に1が3つ以上現れる番号(1は連続していなくて良い)

```python
grep("(1[^1]*){3,}",変数IDs,perl=TRUE,value=TRUE)
```

###週次課題8.2

```python
respon <- read.csv("/content/respon.csv", header=TRUE)
```

```python
data <- respon$解答
```

```python
respon_data_All <- gsub("[^\\d{2}\\.\\d*]","",data,perl=TRUE)
```

```python
respon_data_All
```

```python
respon_data_num <- as.numeric(respon_data_All)
```

```python
respon_data_true <- respon_data_num[!is.na(respon_data_num)]
```

```python
respon_data_true
```

```python
respon_data <- c()
```

```python
for(i in 1 : length(respon_data_true)){
  if(respon_data_true[i] >= 40 && respon_data_true[i] <= 60) respon_data <- c(respon_data,respon_data_true[i])
}
```

```python
hist(respon_data,breaks=seq(40,60,2),main="Histogram")
```

###週次課題8.3

```python
word <- read.csv("/content/web2.txt", header=FALSE)
```

```python
words <- word$V1
```

```python
grep("^[A-Za-z]$",words,perl=TRUE,value=TRUE)
```

a i u e o をこの順に含む単語は?

```python
grep("^[^(i|u|e|o)]*a[^(i|u|e|o)]*i[^(a|u|e|o)]*u[^(a|i|e|o)]*e[^(a|i|u|o)]*o[^(a|i|u|e)]*$",words,perl=TRUE,value=TRUE)
```

aを5つ以上含む単語は?(量化子を使用すること)

```python
grep("^.{1}$",words,perl=TRUE,value=TRUE)
```

```python
grep("(a[^a]*){5,}",words,perl=TRUE,value=TRUE)
```

aまたはAで始まりzで終わる単語は？

```python
grep("^(A|a).*z$",words,perl=TRUE,value=TRUE)
```

同じ文字が3文字以上連続して現れる単語は?

```python
grep("(.)\\1{2,}",words,perl=TRUE,value=TRUE)
```
