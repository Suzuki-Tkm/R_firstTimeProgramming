### Ex.1-7  NE 専門科目の「講義内容」取得
##### 専門科目全体の「講義内容」部分のシラバスデータを科目ごとに取得し，テキストファイルと
して保存
- 科目一覧「subj_NE2022.csv」をセッションストレージに事前に保存
- セッションストレージ内（デフォルトの作業ディレクトリ配下）にサブディレクトリ「tmpd」を作成
- 各科目のファイル名は「科目コード.txt」
- シラバス全科目のデータファイルをtmpdに保存
- tmpdをzip圧縮して（手動）ダウンロード

######参考：「ファイル操作関数」　https://stats.biopapyrus.jp/r/basic/file.html

```python
# [1]
getstxt <- function(url){
  library(rvest)

  dat <- read_html(url)			# 読み込み
  XPath <- "//tr/td[@id='main']/div[@class='contents_1']/form[@id='form']/div[@class='item'][2]/table[@class='syllabus_detail']/tr[1]/td[3]/table/tr/td"				# 「講義内容」部分のXPath
  lec <- html_nodes(dat,xpath=XPath)			# ノード（要素・タグ）の抽出
  lec <- html_text(lec)			# ノードからテキストの抽出
  lec <- gsub("[[:space:]]","",lec)				# 無駄な文字列（空白文字）の削除　※英文等では問題

  return(lec)
}
```

```python
# [2]
savestxt <- function(ids,urls,sdir){

  library(rvest)
  dir.create(sdir)
  for(i in 1:length(ids)){
    fn <- paste0(ids[i],".txt")
    writeLines(getstxt(urls[i]),paste0(sdir,"/",fn))
  }
}
```

```python
# [3]
data <- read.csv("/content/subj_NE2022.csv",header=TRUE)
ids <- data$id
urls <- data$spage

savestxt(ids,urls,"tmpd")
```
