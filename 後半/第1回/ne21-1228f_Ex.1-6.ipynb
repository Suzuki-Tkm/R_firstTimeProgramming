```python
library(rvest)
id <- 32140         # 科目コード
url <- "https://syllabus.acc.senshu-u.ac.jp/syllsenshu/slbssbdr.do?value(risyunen)=2022&value(semekikn)=1&value(kougicd)=32140&value(crclumcd)=606100"				# 科目シラバスのURL（文字列）※手入力で良い
dat <- read_html(url)			# 読み込み
dat						# 確認（DOM形式）
XPath <- "//tr/td[@id='main']/div[@class='contents_1']/form[@id='form']/div[@class='item'][2]/table[@class='syllabus_detail']/tr[1]/td[3]/table/tr/td"				# 「講義内容」部分のXPath
lec <- html_nodes(dat,xpath=XPath)			# ノード（要素・タグ）の抽出
lec						      # 確認
lec <- html_text(lec)			# ノードからテキストの抽出
lec						      # 確認
lec <- gsub("[[:space:]]","",lec)				# 無駄な文字列（空白文字）の削除　※英文等では問題
#lec						      # 確認
fn <- paste0(id,".txt")			    # idからファイルパス（文字列）作成
write(lec, file=fn)			    # ファイル出力
```
