### Ex.1-2　「食べログ」からのデータ取得

```python
# [1]夕食時の予算の「上限」

library(rvest)
html <- read_html("https://tabelog.com/tokyo/A1310/A131002/13167573/") # HTML読み込み
#html						# 確認　※DOM（Document Object Model）形式で表示
tmp <- html_nodes(html, xpath="//a[@class='rdheader-budget__price-target']")  # HTMLの要素抽出
#tmp						  # 確認（2要素）
price <- html_text(tmp[1])		# htmlから属性，テキスト，タグ名を抽出
#price						# 確認
price1 <- gsub("^.","",price)	# 先頭1文字を空文字で置換（＝削除）して
price1 <- gsub(".{7}","",price1)	# 「～」までを削除して
price1 <- gsub(",","",price1)	# カンマを削除して
price1 <- as.numeric(price1)		# 数値化
price1					# 確認
```

```python
# [2]クチコミ評価（★の数値）

library(rvest)
html <- read_html("https://tabelog.com/tokyo/A1310/A131002/13167573/") # HTML読み込み
#html						# 確認　※DOM（Document Object Model）形式で表示
tmp <- html_nodes(html, xpath="//span[@class='rdheader-rating__score-val-dtl']")  # HTMLの要素抽出
#tmp						  # 確認（2要素）
price <- html_text(tmp)		# htmlから属性，テキスト，タグ名を抽出
price <- as.numeric(price)		# 数値化
price					# 確認
```
