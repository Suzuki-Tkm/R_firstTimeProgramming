自作関数

```python
svar <- function(x) {
  #(1) x(ベクトル)から平均 m(スカラー)を求める.
  m <- mean(x); #m
  #(2) x, m から偏差 d(ベクトル)を求める.
  d <- m-x; #d
  #(3) d から偏差平方 d2(ベクトル)を求める.
  d2 <- d^2; #d2
  #(4) d2 から偏差平方和 ss(スカラー)を求める.
  ss <- sum(d2); #ss
  #(5) x からデータ数 n(スカラー)を求める.
  n <- length(x); #n
  #(6) ss, n から標本分散 res(スカラー)を求める.
  res <- 1/n*ss; #res

  return (res)
}
```

```python
x <- 1:5
svar(x)
```
