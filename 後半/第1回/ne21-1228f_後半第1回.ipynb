### 例題.1-1　食事の予算
##### 食べログのページから，お昼の最低予算(数値)を抽出（rvestパッケージ使用）  
まずHTMLファイルの内容・構造を把握しておくこと

```
library(rvest)
html <- read_html("https://tabelog.com/tokyo/A1310/A131002/13167573/") # HTML読み込み
html						# 確認　※DOM（Document Object Model）形式で表示
tmp <- html_nodes(html, xpath="//a[@class='rdheader-budget__price-target']")  # HTMLの要素抽出
#tmp <- html_nodes(html, xpath="//p[2]/span/a")  # これもOK（シンプルだが分かりにくいかも）
tmp						  # 確認（2要素）
price <- html_text(tmp[2])		# htmlから属性，テキスト，タグ名を抽出
#price <- html_text(tmp)		# シンプルな方
price						# 確認
price1 <- gsub("^.",　"",　price)	# 先頭1文字を空文字で置換（＝削除）して
price1 <- gsub("～.*",　"",　price1)	# 「～」以降を削除して
price1 <- gsub(",",　"",　price1)	# カンマを削除して
price1 <- as.numeric(price1)		# 数値化
price1					# 確認
# 参考：1行で求めるとすれば...
# as.numeric(gsub("^.(\\d*?)(,?)(\\d+)～.*", "\\1\\3", price, perl=T))
# gsubはsubでもOK．以下でも（おそらく）．
```

### 例題.1-2　食事の予算（パイプ処理）
##### 参照：『R言語でパイプ処理を利用してコードを見やすく』  
http://webbeginner.hatenablog.com/entry/2015/12/29/211159

```
library(rvest)
html <- read_html("https://tabelog.com/tokyo/A1310/A131002/13167573/") # HTML読み込み
html	# 確認　※DOM（Document Object Model）形式で表示
tmp <- html_nodes(html, xpath="//a[@class='rdheader-budget__price-target']")  # HTMLの要素抽出
tmp					# 確認（2要素）
price <- html_text(tmp[2])			# htmlから属性，テキスト，タグ名を抽出
price					# 確認
price1 <- price %>% gsub(pattern="^.",replacement="") %>%
  gsub(pattern="～.*",replacement="") %>% gsub(pattern=",",replacement="") %>% as.numeric
price1					# 確認
```

### 例題.1-3　zip圧縮  
1. セッションストレージ内にディレクトリ「testd」を作成し，適当なデータファイルを入れる．
1. ディレクトリを圧縮して「testd.zip」を作成し，ローカル（PC内）に手動ダウンロードして，内容を確認．
1. ディレクトリ「testd」を手動で削除（※最初に中のファイルを削除しておくこと）．
1. セッションストレージ内でzipファイルを展開し，内容を確認．

##### 参考：Rのファイル操作関数  https://stats.biopapyrus.jp/r/basic/file.html

```
# [1]
dir.create("testd")        # ディレクトリ「testd」作成
item <- c("A", "B", "C")   # データ作成
vol <- c(10, 20, 30)
write.table(item, "testd/item.csv", row.names=FALSE, col.names=FALSE)    # データ保存
write.table(vol, "testd/vol.csv", row.names=FALSE, col.names=FALSE)
```

```
# [2]
system("apt -y install zip && zip -r testd.zip testd")    # ディレクトリ圧縮
#「testd.zip」をローカルに手動ダウンロードし解凍して内容確認のこと
# [3] ディレクトリ「testd」を手動で削除（中のファイルを最初に全削除）
```

```
# [4]
system("apt -y install zip && unzip testd.zip")    # zip解凍（ディレクトリ生成を確認）
```
