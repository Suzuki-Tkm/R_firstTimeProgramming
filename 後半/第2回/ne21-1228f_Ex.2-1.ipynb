```python
# [2] d：データフレーム（の列要素）→ 形態素
ID <- 1:3;ID    # ベクトル{1,2,3}をオブジェクト（変数）IDに代入
Sex <- c("F", "M", "F")    # ベクトル{"F", "M", "F"}をオブジェクト（変数）Sexに代入
Rem <- c("今朝は早く起きた。", "昨夜は遅く寝た。", "今朝の食事は遅かった．")    # ベクトル{"今朝は早く起きた。", "昨夜は遅く寝た。", "今朝の食事は遅かった．"}をオブジェクト（変数）Remに代入
d <- data.frame(ID, Sex, Rem)    # 引数にベクトルを複数並べ(ID,Sex,Rem)、そのベクトルを列にもつデータフレームを作成し、オブジェクト（変数）dに代入。
d    # オブジェクト（変数）dを出力。
```

```python
res <- docDF(d, 3, type=1); res    # 第1引数で指定されたデータフレーム(d)のNgram行列、あるいはターム・文書行列を作成する。第2引数で列(3列)を指定している。typeについては下記参照。
                                    #以上をオブジェクト（変数）resに代入。resを出力。
res <- docDF(d, 3, type=1, pos=c("名詞","形容詞")); res    #第1引数で指定されたデータフレーム(d)のNgram行列、あるいはターム・文書行列を作成する。第2引数で列(3列)を指定している。
                                                            #posはc("名詞", "形容詞")を指定し抽出する。typeについては下記参照。
                                                            #以上をオブジェクト（変数）resに代入。resを出力。
res <- docDF(d, 3, type=1, pos=c("名詞","形容詞"), minFreq=2); res   #第1引数で指定されたデータフレーム(d)のNgram行列、あるいはターム・文書行列を作成する。第2引数で列(3列)を指定している。
                                                                       #posはc("名詞", "形容詞")を指定し抽出する。
                                                                       #minFreqは全ファイルを通じての合計頻度(2)の閾値を指定し抽出する。typeについては下記参照。
                                                                       #以上をオブジェクト（変数）resに代入。resを出力。
#typeについては0は文字のGramを１はタームのGramを作成する。※Gramとはテキスト分割方法。
```

```python
# [3]　dn：（文章ファイルの）ディレクトリ → 形態素（文書ターム行列）
dn <- "tmpd"    # 文字列をオブジェクト（変数）dnに代入
dir.create(dn)    # ディレクトリ(dn)作成
fn <- c("f1.txt", "f2.txt", "f3.txt")    # # ベクトル{"f1.txt", "f2.txt", "f3.txt"}をオブジェクト（変数）fnに代入
fn <- paste0(dn, "/", fn)    # ()の中の要素(dn,"/",fn)を文字型に変えて, 繋げて出力する。
for (i in 1:3){    # 繰り返し処理(iが1から足していき3まで)を行う。処理は{から始まる。
  write(Rem[i], fn[i])    # 文字列Rem[i](Remのi番目)をテキストファイル(fn[i](fnのi番目))に出力。
}    # }までを繰り返し実行する。
dir(dn)    # カレントディレクトリ(“/content/tmpd”)のディレクトリ名、ファイル名が出力。※引数は/content以降のパスを指定。
# G.Colab.の左の「ファイル」メニューを開き「更新」アイコンを押して作ったディレクトリを確認せよ
```

```python
res <- docDF(dn, type=1); res    #第1引数で指定されたファイル(dn)のNgram行列、あるいはターム・文書行列を作成する。typeについては下記参照。
                                  #以上をオブジェクト（変数）resに代入。resを出力。
res <- docDF(dn, type=1, minFreq = 2); res    #第1引数で指定されたファイル(dn)のNgram行列、あるいはターム・文書行列を作成する。typeについては下記参照。
                                                #minFreqは全ファイルを通じての合計頻度(2)の閾値を指定し抽出する。
                                                #以上をオブジェクト（変数）resに代入。resを出力。
res <- docDF(dn, type=1, pos=c("名詞", "動詞")); res    #第1引数で指定されたファイル(dn)のNgram行列、あるいはターム・文書行列を作成する。typeについては下記参照。
                                                         #posはc("名詞", "形容詞")を指定し抽出する。
                                                         #以上をオブジェクト（変数）resに代入。resを出力。
#typeについては0は文字のGramを１はタームのGramを作成する。※Gramとはテキスト分割方法。
```

```python
# [4]　dn：ディレクトリ → 形態素（文書ターム行列）※重み付き　
# 参照：http://www.ic.daito.ac.jp/~mizutani/mining/term_document_matrix.html
# 「検索語・文書行列の重み付け 〜 TF・IDF法」と
# 「TF・IDF法による検索語・文書行列の規格化」の部分を読むこと
res <- docDF(dn, type=1, pos=c("名詞", "形容詞", "動詞"), weight="tf*idf"); res
    #第1引数で指定されたファイル(dn)のNgram行列、あるいはターム・文書行列を作成する。typeについては下記参照。
    #posはc("名詞", "形容詞", "動詞")を指定し抽出する。
    #weightは"tf*idf*norm" などの重みを指定可能にする。
    #tfは1つの文書内における出現頻度。
    #idfは全ての文書の内いくつの文書にその単語が出現しているのかの割合の逆数の対数を取ったもの。
    #以上をオブジェクト（変数）resに代入。resを出力。
res <- docDF(dn, type=1, pos=c("名詞", "形容詞", "動詞"), weight="tf*idf*norm"); res
    #第1引数で指定されたファイル(dn)のNgram行列、あるいはターム・文書行列を作成する。typeについては下記参照。
    #posはc("名詞", "形容詞", "動詞")を指定し抽出する。
    #weightは"tf*idf*norm" などの重みを指定可能にする。
    #tfは1つの文書内における出現頻度。
    #idfは全ての文書の内いくつの文書にその単語が出現しているのかの割合の逆数の対数を取ったもの。
    #normは正規化を行う。※各列の要素を2乗して合計すると1になる。
    #以上をオブジェクト（変数）resに代入。resを出力。
res[,4:6]    # data.frame(res)4～6列を出力
colSums(res[,4:6]^2)  #data.frame(res)4～6列の各列を2乗した合計を出力。
#typeについては0は文字のGramを１はタームのGramを作成する。※Gramとはテキスト分割方法。
```

```python
# [5] 　dn：ディレクトリ（or fn：ファイル） → N-gram
# 参考：「Nグラム統計」http://www.ic.daito.ac.jp/~mizutani/mining/n_gram.html
# に目を通すこと
dn <- "tmpd"    #文字列をオブジェクト（変数）dnに代入
res <- NgramDF2(dn, type=1, N=2); res    # 引数で指定されたディレクトリ(dn)について，N-gram を計る。
                                            # typeはtype=0の時は言語単位を文字に、type=1の時は言語単位を形態素に、type=2の時は言語単位を品詞にしてその単位をカウントする。
                                            #　N はnグラムの時のgram を計る。今回は2グラムである。
                                            #以上をオブジェクト（変数）resに代入。resを出力。
res <- NgramDF2(dn, type=1, N=2, pos=c("名詞", "動詞")); res    # 引数で指定されたディレクトリ(dn)について，N-gram を計る。
                                                                   # typeはtype=0の時は言語単位を文字に、type=1の時は言語単位を形態素に、
                                                                   #type=2の時は言語単位を品詞にしてその単位をカウントする。
                                                                   #　N はnグラムの時のgram を計る。今回は2グラムである。
                                                                   #posはc("名詞", "動詞")を指定し抽出する。
                                                                   #以上をオブジェクト（変数）resに代入。resを出力。
res <- NgramDF2(dn, type=1, N=3, pos=c("名詞", "形容詞", "動詞")); res     # 引数で指定されたディレクトリ(dn)について，N-gram を計る。
                                                                             # typeはtype=0の時は言語単位を文字に、type=1の時は言語単位を形態素に、
                                                                             #type=2の時は言語単位を品詞にしてその単位をカウントする。
                                                                             #　N はnグラムの時のgram を計る。今回は3グラムである。
                                                                             #posはc("名詞", "形容詞", "動詞")を指定し抽出する。
                                                                             #以上をオブジェクト（変数）resに代入。resを出力。
res <- NgramDF2(dn, type=2, N=2); res    # 引数で指定されたディレクトリ(dn)について，N-gram を計る。
                                            # typeはtype=0の時は言語単位を文字に、type=1の時は言語単位を形態素に、type=2の時は言語単位を品詞にしてその単位をカウントする。
                                            #N はnグラムの時のgram を計る。今回は2グラムである。
                                            #以上をオブジェクト（変数）resに代入。resを出力。
fn <- "tmpd/f2.txt"    # 文字列をオブジェクト（変数）fnに代入。
res <- NgramDF2(fn, type=0, N=2); res    # 引数で指定されたファイル(fn)について，N-gram を計る。
                                            # typeはtype=0の時は言語単位を文字に、type=1の時は言語単位を形態素に、type=2の時は言語単位を品詞にしてその単位をカウントする。
                                            #N はnグラムの時のgram を計る。今回は2グラムである。
                                            #以上をオブジェクト（変数）resに代入。resを出力。
```

```python
# [6] コロケーションとTスコア，MIスコア
# 参考：「コロケーションと語彙分析」
# http://www.ic.daito.ac.jp/~mizutani/mining/collocation.html
# に目を通すこと 
fn <- "tmpd/f3.txt"    #文字列をオブジェクト（変数）fnに代入。
readLines(fn)    # コネクションからオブジェクト(fn)の全部のテキストラインを読み取る。
collocate(fn, node="は", span=1)    # 指定されたファイル(fn)について，nodeで指定された語(ただし形態素原形)("は")の前後に出てくる単語頻度を計算する。
                                     #spanはnodeから測った前後の語数(1)を示す。
                                     #Before列はノードを含む後方スパンでの出現頻度、After列はノードを含まない前方スパン内の出現頻度.。
collocate(fn, node="は", span=2)    # 指定されたファイル(fn)について，nodeで指定された語(ただし形態素原形)("は")の前後に出てくる単語頻度を計算する。
                                     #spanはnodeから測った前後の語数(2)を示す。
                                     #Before列はノードを含む後方スパンでの出現頻度、After列はノードを含まない前方スパン内の出現頻度.。
collocate(fn, node="は", span=3)    # 指定されたファイル(fn)について，nodeで指定された語(ただし形態素原形)("は")の前後に出てくる単語頻度を計算する。
                                     #spanはnodeから測った前後の語数(3)を示す。
                                     #Before列はノードを含む後方スパンでの出現頻度、After列はノードを含まない前方スパン内の出現頻度.。
```

```python
res <- collocate(fn, node="は", span=2)    # 指定されたファイル(fn)について，nodeで指定された語(ただし形態素原形)("は")の前後に出てくる単語頻度を計算する。
                                            #spanはnodeから測った前後の語数(2)を示す。
                                            #Before列はノードを含む後方スパンでの出現頻度、After列はノードを含まない前方スパン内の出現頻度.。
                                            #以上をオブジェクト（変数）resに代入。

collScores(res, node="は", span=2)    #collocate() 関数の出力から T値, MI値 を返す。
                                       #T値は、語の頻度情報を重視するために高頻度語が高く評価される傾向があり、頻繁につかわれる一般性の高いコロケーションの評価に適しているとされている。
                                       #MI値は、頻度は低いが特殊な結びつきをしているコロケ ーションがうまく検出できるとされている（低頻度の語を強調する傾向がある）。
                                       #指定されたcollocate()関数(res)について，nodeで指定された語(ただし形態素原形)("は")の前後に出てくる単語頻度を計算する。
                                       #spanはnodeから測った前後の語数(2)を示す。
```
